##**1. Amostragem e sua aplicação em dados comunitários**

Nesta aula, vamos entender o que é amostragem, por que ela é importante
e como ela pode ser aplicada em pesquisas com dados geoespaciais e sociais
em comunidades vulnerabilizadas.

Em muitos contextos, não é possível entrevistar toda a população de um território. Por isso, a amostragem ajuda a definir quem será incluído na pesquisa sem perder a capacidade de compreender a realidade local.

----

## **2. O que é Amostragem?**

Amostragem é o processo de selecionar uma parte da população para representar
o todo. Em vez de entrevistar todas as famílias de uma comunidade, o pesquisador
pode selecionar uma parte delas e, a partir disso, produzir análises.

Em alguns projetos, a amostragem pode ser utilizada para:

- levantamento socioeconômico
- análise de condições territoriais
- produção de dados geoespaciais em contextos onde o acesso a toda a população
  é difícil
- Estudos em que o tempo, os recursos e a equipe são limitados

## **3. População x amostra**

**População** é o conjunto total de elementos que queremos estudar.

- Exemplo: Todas as famílias vinculadas à uma Escola comunitaria.

**Amostra** é uma parte dessa população.

- Exemplo: 50 famílias selecionadas para representar o conjunto total.

Uma boa amostra precisa ser planejada para reduzir distorções
e evitar conclusões enviesadas.

---

## **4. Alguns Tipos de amostragem**

##### **1 - Amostragem aleatória simples:**
Cada unidade da população tem a mesma chance de ser escolhida.

##### **2 - Amostragem estratificada:**
A população é dividida em grupos e a amostra é retirada de cada grupo.

- Exemplo: separar famílias por setor da comunidade, faixa de renda ou presença de crianças em diferentes séries.

##### **3 - Amostragem por conglomerados:**
Selecionam-se grupos inteiros em vez de indivíduos isolados.

- Exemplo: selecionar determinadas ruas, becos ou setores da comunidade para a coleta.

##### **4 - Amostragem por conveniência**
A seleção é feita com base na facilidade de acesso.

----

## **5. Ligação com projetos de pesquisa comunitarios**

**No Exemplo do projeto de pesquisa:** avaliação de impacto de processos de geração cidadã de dados geoespaciais em comunidades vulnerabilizadas, a amostragem é importante porque:

1. ajuda a planejar coletas quando não é possível alcançar toda a população;
2. influencia diretamente a qualidade e a representatividade dos dados;
3. pode afetar a leitura espacial do território, especialmente quando alguns
   setores ficam de fora;
4. precisa considerar desigualdades internas da comunidade;
5. interfere na capacidade de produzir diagnósticos e apoiar decisões locais.

Em pesquisas comunitárias, a amostra não deve ser pensada apenas como um cálculo
estatístico. Ela também precisa dialogar com o território, com a organização local e com os grupos menos visíveis da comunidade.

## **6. Exemplos com alguns Tipos de Amostragem:**

```python
amostra_aleatoria = df.sample(40, random_state=42)

display(Markdown("""
### **6.1 - Exemplo de amostragem aleatória simples**

Aqui selecionamos 40 famílias ao acaso de uma determinada comunidade.
Esse método é simples, mas pode deixar grupos importantes sub-representados
dependendo da distribuição da comunidade.
"""))

display(amostra_aleatoria.head())
```

```python
amostra_estratificada = (
    df.groupby("setor", group_keys=False)
      .apply(lambda x: x.sample(max(5, int(len(x) * 0.20)), random_state=42))
      .reset_index(drop=True)
)

display(Markdown("""
### **6.2 - Exemplo de amostragem estratificada**

Agora a amostra é retirada dentro de cada setor.
Isso ajuda a garantir que diferentes partes da comunidade estejam representadas.

 - Em pesquisas territoriais, essa costuma ser uma escolha mais adequada do que a amostragem aleatória simples, principalmente quando há desigualdades espaciais.
"""))

display(amostra_estratificada.head())
```

```python
setores_sorteados = np.random.choice(df["setor"].unique(), size=2, replace=False)
amostra_conglomerado = df[df["setor"].isin(setores_sorteados)]

display(Markdown(f"""
### **6.3 - Exemplo de amostragem por conglomerados**

Aqui foram sorteados dois setores inteiros da comunidade: {", ".join(setores_sorteados)}.

- Esse tipo de amostragem pode ser útil quando a coleta em campo é feita por áreas,
ruas, setores ou blocos territoriais.
"""))

display(amostra_conglomerado.head())
```

## **7. A que conclusão chegemos a respeito do uso da Amostragem?**

A escolha da amostra pode mudar os resultados da pesquisa. Se alguns setores forem mais acessíveis do que outros, a coleta pode concentrar
dados em certas áreas e invisibilizar outras.

No caso do projeto de pesquisa citado como exemplo, isso é ainda mais importante porque o tema envolve comunidades vulnerabilizadas. Uma amostra mal planejada pode reforçar desigualdades, deixando de fora grupos menos visíveis ou áreas mais periféricas. Por isso, a decisão metodológica sobre a amostragem precisa ser pensada junto com o conhecimento do território.
